## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup
url='https://www.dcard.tw/_api/forums/pet/posts?popular=true'

r=requests.get(url)
r.encoding='utf-8'
response=r.text
soup=BeautifulSoup(response,'html5lib')
print(soup)
print(type(soup))
```

## Grab + PyQuery

```python

print('requests + BeautifulSoup')
print('這個方式在response時收到的資料,可以以UTF-8的文字格式呈現,而資料的型態是bs4.BeautifulSoup')

print('Grab + PyQuery')
print('這個方式在response時收到的資料,使用encoding呈現的資料仍為亂碼,出現這個的情況原因不明,資料的型態是pyquery.pyquery.PyQuery,與requests+BeautifulSoup不同格式')
```
