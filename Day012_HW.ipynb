### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

words=[]

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
    words.append([d.find(class_="date").text,d.find('em').text,d.find_all('a')[-1].text])
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
columns=['日期','類別','標題']

table=pd.DataFrame(words,columns=columns)
table
```

### ③ 哪一個類別的文章最多

```python

group=table.groupby('類別')
table2=pd.DataFrame(group.size(),columns=['文章數量'])
table2.sort_values(by = '文章數量',ascending=False)
```
