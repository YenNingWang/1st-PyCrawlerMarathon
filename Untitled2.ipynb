```python
from selenium import webdriver
from bs4 import BeautifulSoup
driver = webdriver.Chrome('C:/Windows/chromedriver.exe')
driver.get('https://www.cupoy.com/newsfeed/topicgrp/tech_tw')
```

```python
import time
time.sleep(5)
```

```python
# 設定list與迴圈
count = 1
title500 = []
category_list = [] # 先把類別放進來
category_count = {} # 整理各類別有多少個

while (count <= 500):
    html_source = driver.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    news = soup.find('div', class_ = 'ReactVirtualized__Grid__innerScrollContainer').find_all(class_='sc-eEieub sc-iuDHTM ibJqYc')
    
    for titles in news:
        category = titles.find(class_ = 'sc-gacfCG bPSpUf').text
        title = titles.find(class_ = 'sc-erNlkL sc-ekulBa hDLssh').text
         # 防止爬到重複的文章
        if title not in title500:
            print(count, category, title)
            count += 1 # count表文章爬到的數目
            
        # 將各資料存進list以供後續分析
        title500.append({
        '類別': category,
        '標題': title
        })
        category_list.append(category)
        # 停止條件，抓到500篇文章即終止程式
        if count == 501:
            break
            
# 關閉瀏覽器
driver.quit();
```

```python
# 以表格顯示
import pandas as pd
import numpy as np

pd.DataFrame(title500)
```

```python
# 總計類別，並選出最多類別
for category in category_list:
    if category not in category_count:
        category_count[category] = 1
    else:
        category_count[category] += 1

print('類別:', category_count)
print('類別最多為:', max(category_count, key = category_count.get))
```
